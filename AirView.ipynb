```python
# AirView
# plot air voltage signals from a CSV file
# with Matplotlib
#
# å°ç£åœ°éœ‡é æ¸¬ç ”ç©¶æ‰€ æ‰€é•·
# æž—æ¹§æ£®
# 2017-11-17 09:52 UTC+8

%matplotlib notebook
import matplotlib
import pandas as pd

from matplotlib.figure import Figure
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import matplotlib.font_manager as fm

graph_title = 'AirView 5.0.5 å®œè˜­ç«™ Yilan Station ç©ºæ°£2è™Ÿ Air 2 (Arduino Uno + LF298N)'
csv_file_name = '2017-05-16 AirView.csv'
df = pd.read_csv(csv_file_name, names=['Time', 'Air Voltage (mV)'], 
                 parse_dates=['Time'])
# set x and y
x = df['Time']
y = df['Air Voltage (mV)']


# è¨­å®šåœ–å½¢å°ºå¯¸èˆ‡å“è³ª
figure = plt.figure(figsize=(10, 5), dpi=100)
axis1 = plt.axes()

#figure = Figure(figsize=(10, 5), dpi=100)
#axis1 = figure.add_subplot(111)
# ç¹ªè£½åœ–å½¢
axis1.plot(x, y)
#axis1.plot(x, y, color='black')
axis1.set_title(graph_title, fontproperties='SimHei', fontsize=14)
#axis1.set_title(graph_title, fontproperties='SimHei', fontsize=24)
axis1.set_xlabel('æ™‚é–“ Time', fontproperties='SimHei')
axis1.set_ylabel('ç©ºæ°£é›»å£“ Air Voltage (mV)', fontproperties='SimHei')

# Release memory
import gc
del df
gc.collect()
```
